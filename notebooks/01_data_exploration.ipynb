```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
pd.set_option('display.float_format', '{:.2f}'.format)
print('All imports successful!')
```

####  Load all datasets

```python
# Cell 2: Load all CSV files
orders = pd.read_csv('../data/olist_orders_dataset.csv')
items = pd.read_csv('../data/olist_order_items_dataset.csv')
payments = pd.read_csv('../data/olist_order_payments_dataset.csv')
reviews = pd.read_csv('../data/olist_order_reviews_dataset.csv')
customers= pd.read_csv('../data/olist_customers_dataset.csv')
products = pd.read_csv('../data/olist_products_dataset.csv')
sellers = pd.read_csv('../data/olist_sellers_dataset.csv')
geo = pd.read_csv('../data/olist_geolocation_dataset.csv')
cat_map = pd.read_csv('../data/product_category_name_translation.csv')
# Print summary
datasets = {'orders': orders, 'items': items, 'payments': payments,
 'reviews': reviews, 'customers': customers, 'products': products,
 'sellers': sellers, 'geolocation': geo, 'category_map': cat_map}
for name, df in datasets.items():
 print(f'{name:20s}: {df.shape[0]:>8,} rows x {df.shape[1]} columns')
```

 #### Check for nulls and data types

```python
# Cell 3: Data quality check for orders table
print('=== ORDERS TABLE ===')
print(f'Shape: {orders.shape}')
print()
print('NULL COUNTS:')
print(orders.isnull().sum())
print()
print('DATA TYPES:')
print(orders.dtypes)
print()
print('DUPLICATE ROWS:', orders.duplicated().sum())
print()
print('ORDER STATUS DISTRIBUTION:')
print(orders['order_status'].value_counts())
print()
print('DATE RANGE:')
print('Earliest order:', orders['order_purchase_timestamp'].min())
print('Latest order: ', orders['order_purchase_timestamp'].max())
```

 #### Run the same quality check on order_items

```python
# Cell 4: Items + payments quality check
print('=== ORDER ITEMS ===')
print(items.isnull().sum())
print('Price stats:')
print(items['price'].describe())
print()
print('=== PAYMENTS ===')
print(payments.isnull().sum())
print('Payment types:')
print(payments['payment_type'].value_counts())
print()
print('=== REVIEWS ===')
print(reviews.isnull().sum())
print('Score distribution:')
print(reviews['review_score'].value_counts().sort_index())
```
